```python

import pandas as pd
import numpy as np

pd.set_option("display.max_columns", 200)
pd.set_option("display.max_rows", 200)

DATA_PATH = "matchData.csv"

df = pd.read_csv(DATA_PATH, low_memory=False)
print("Loaded rows:", len(df))

# Quick preview
df.head()
```

```python
# Convert team wins to numeric (0/1)
df["team0Win"] = df["team0Win"].astype(int)
df["team1Win"] = df["team1Win"].astype(int)

# Create unified win column representing team0 perspective
df["win"] = df["team0Win"]

# Extract major.minor patch (e.g. '15.16')
df["patch"] = df["gameVersion"].str.extract(r"^(\d+\.\d+)")

# Verify columns
print("Columns:", len(df.columns))
df.head()
```

```python
# Should always be 10 participants
participant_champ_cols = [f"participant{i}ChampionName" for i in range(10)]
missing_names = df[participant_champ_cols].isna().sum().sum()

print("Missing participant champion names:", missing_names)

# Check ban columns exist
ban_cols = [f"team{t}Ban{i}ChampionId" for t in (0,1) for i in range(5)]
missing_bans = df[ban_cols].isna().sum().sum()
print("Missing ban IDs:", missing_bans)

# Check durations
print("Game duration mean:", df["gameDuration"].mean())
print("Game duration distribution:")
df["gameDuration"].describe()
```

```python
import numpy as np
import pandas as pd

df_step3 = df_clean.copy()

# Raw ban ID columns
ban_id_cols = [f"team{t}Ban{i}ChampionId" for t in (0,1) for i in range(5)]

# New (clean) ban name columns to create
ban_name_cols = [c.replace("ChampionId", "ChampionName_clean") for c in ban_id_cols]

print("Ban ID cols:", ban_id_cols)
print("New ban NAME cols:", ban_name_cols)
```

```python
import pandas as pd

# start from cleaned df (after useless column drops)
df3 = df.copy()

# collect (gameVersion, championId, championName) from all 10 slots
rows = []
for i in range(10):
    tmp = df3[[
        "gameVersion",
        f"participant{i}ChampionId",
        f"participant{i}ChampionName"
    ]].dropna()

    tmp = tmp.rename(columns={
        f"participant{i}ChampionId": "championId",
        f"participant{i}ChampionName": "championName"
    })
    rows.append(tmp)

champ_map = pd.concat(rows, ignore_index=True)

# drop exact duplicates
champ_map = champ_map.drop_duplicates()

# check if any (patch, id) maps to multiple names
check = champ_map.groupby(
    ["gameVersion", "championId"]
)["championName"].nunique().sort_values(ascending=False)

print("Max different names per (patch, id):", check.iloc[0])
print("Problematic entries (if any):")
print(check[check > 1].head())
```

```python
# choose the most frequent name per (patch, id) just in case
champ_map = (
    champ_map
    .groupby(["gameVersion", "championId", "championName"])
    .size()
    .reset_index(name="count")
    .sort_values(["gameVersion", "championId", "count"], ascending=[True, True, False])
    .drop_duplicates(subset=["gameVersion", "championId"])
    .drop(columns="count")
)
print("Final champ_map size:", len(champ_map))
champ_map.head()
```

```python
# ban ID columns
ban_id_cols = [f"team{t}Ban{i}ChampionId" for t in (0, 1) for i in range(5)]

# make a fresh working copy with match_index
df3 = df.copy().reset_index(drop=False).rename(columns={"index": "match_index"})

# long-format bans
ban_long = df3[["match_index", "gameVersion"] + ban_id_cols].melt(
    id_vars=["match_index", "gameVersion"],
    value_vars=ban_id_cols,
    var_name="ban_slot",
    value_name="championId"
)

# merge with (patch, id) â†’ name mapping
ban_long = ban_long.merge(
    champ_map,   # from previous step
    on=["gameVersion", "championId"],
    how="left"
)

# pivot back: one column per ban slot, values = championName
ban_name_wide = ban_long.pivot(
    index="match_index",
    columns="ban_slot",
    values="championName"
)

# rename columns idâ†’name
ban_name_wide = ban_name_wide.rename(
    columns=lambda c: c.replace("ChampionId", "ChampionName")
)

# drop any existing columns with the same names before joining
cols_to_drop = [c for c in ban_name_wide.columns if c in df3.columns]
if cols_to_drop:
    df3 = df3.drop(columns=cols_to_drop)

df3 = df3.join(ban_name_wide, on="match_index")

# drop helper index
df3 = df3.drop(columns=["match_index"])
df3.head()
```

```python
# detect ban name columns (all after the mapping step)
ban_name_cols = [c for c in df3.columns if "Ban" in c and "ChampionName" in c]

print("Ban name columns:", ban_name_cols)

# count all bans across all games
ban_name_counts = df3[ban_name_cols].stack().value_counts(dropna=True)

print("\nTop 20 most banned champions:")
ban_name_counts.head(20)
```
